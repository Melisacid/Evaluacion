```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl 
```

```python
#Carga desde un archivo .xlsx sin indice
df_eua= pd.read_csv('data_clean_california_eua.csv') 
df_esp= pd.read_csv('data_clean_girona_españa.csv') 
df_mex= pd.read_csv('data_clean_df_mexico.csv') 
```

```python
tf_df_eua=freq_tbl(df_eua)
```

```python
#verificamos nulos de los frames
df_mex=df_mex.fillna(method="ffill") 
df_mex.isnull().sum().sum()
df_esp=df_esp.fillna(method="ffill") 
df_esp.isnull().sum().sum()

nulos_eua=df_eua.isnull().sum().sum()
nulos_esp=df_esp.isnull().sum().sum()
nulos_mex=df_mex.isnull().sum().sum()

print(nulos_eua)
print(nulos_esp)
print(nulos_mex)
```

```python
#Obtengo un análisis univariado de una variable categórica en específico
var1_eua= freq_tbl(df_eua['host_identity_verified'])
var1_eua= var1_eua[var1_eua['frequency']>1]
var1_eua= var1_eua.set_index('host_identity_verified')

var1_esp=freq_tbl(df_esp['host_identity_verified'])
var1_esp= var1_esp[var1_esp['frequency']>1]
var1_esp= var1_esp.set_index('host_identity_verified')

var1_mex=freq_tbl(df_mex['host_identity_verified'])
var1_mex= var1_mex[var1_mex['frequency']>1]
var1_mex= var1_mex.set_index('host_identity_verified')

print("variable 1 = EUA-host_identity_verified")
print(var1_eua)
print("\n")
print("variable 1 = ESP-host_identity_verified")
print(var1_esp)
print("\n")
print("variable 1 = MEX-host_identity_verified")
print(var1_mex)
```

```python
var2_eua = freq_tbl(df_eua['review_scores_location'])
var2_eua = var2_eua[var2_eua['frequency'] > 2]
var2_eua = var2_eua.set_index('review_scores_location')

var2_esp = freq_tbl(df_esp['review_scores_location'])
var2_esp = var2_esp[var2_esp['frequency'] > 2]
var2_esp = var2_esp.set_index('review_scores_location')

var2_mex = freq_tbl(df_mex['review_scores_location'])
var2_mex = var2_mex[var2_mex['frequency'] > 2]
var2_mex = var2_mex.set_index('review_scores_location')

print("variable 2 = EUA-review_scores_location")
print(var2_eua)
print("\n")
print("variable 2 = ESP-review_scores_location")
print(var2_esp)
print("\n")
print("variable 2 = MEX-review_scores_location")
print(var2_mex)
```
